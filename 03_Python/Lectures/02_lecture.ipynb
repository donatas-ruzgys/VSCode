Užduotyse naudojami list, dict, for, while, if, elif, else, break, continue, taip pat sąrašų ir žodynų metodai kaip pop(), remove(), reverse(), append(), items(), type().

```python
# 1. Sukurkite sąrašą su skaičiais nuo 1 iki 10. Panaudokite `for` ciklą ir atspausdinkite tik lyginius skaičius. Naudokite `continue`.
first_list = list(range(1, 11))

for number in first_list:
    if number % 2 != 0:
        continue
    print(number)
```

```python
# 2. Turite sąrašą: ["apple", "banana", "cherry", "date"]. Naudokite `for` ciklą ir `if`, kad atspausdintumėte vaisius, kurių ilgis didesnis nei 5.
vaisiai = ["apple", "banana", "cherry", "date"]

for vaisius in vaisiai:
    if len(vaisius) > 5:
        print(vaisius)
```
