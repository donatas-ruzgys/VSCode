```python
# 1. Sukurkite sąrašą su 5 parduotų produktų pavadinimais (list) ir atspausdinkite pirmą ir paskutinį naudodami indeksus.
# 1. Create a list containing the names of 5 sold products and print the first and last items using indexes.

products = ["Milk", "Bread", "Cheese", "Apples", "Butter"]

print(products[0])
print(products[-1])
```

```python
# 2. Sukurkite tuple su klientų vardais. Paverskite jį į sąrašą, pridėkite naują vardą ir vėl konvertuokite į tuple.
# 2. Create a tuple containing customer names. Convert it into a list, add a new name, and then convert it back into a tuple.

customers = ("Ana", "Manul", "Mindaugas")
print(customers)
customers_list = list(customers)
print(customers_list)
customers_list.append("Donatas")
print(customers_list)
customers = tuple(customers_list)
print(customers)
```

```python
# 3. Turite sąrašą kainų [10, 20, 15, 30]. Konvertuokite į set ir parodykite, kiek yra unikalių kainų. Naudokite set() ir len().
# 3. You have a list of prices [10, 20, 15, 30]. Convert it into a set and show how many unique prices there are. Use set() and len().

prices = [10, 20, 15, 30]
prices_set = set(prices)
print(f"Count of unique prices: {len(prices_set)}")
```

```python
# 4. Sukurkite žodyną (dict) su produkto pavadinimu kaip raktu ir pardavimų skaičiumi kaip reikšme. Pakeiskite vieno produkto reikšmę. 
# Naudokite dict indeksavimą.

# 4. Create a dictionary (dict) with the product name as the key and the number of sales as the value. 
# Update the value of one product using dictionary indexing.

sales = {
    "Bread": 10,
    "Milk": 20,
    "Cheese": 6
}
print(sales)
sales["Milk"] = 18
print(f"After update price: {sales}")
```

```python
# 5. Turite list = ["A", "B", "C"] ir set = {"B", "C", "D"}. Raskite bendrus elementus (naudoti set operaciją: intersection).
# 5. You have a list = ["A", "B", "C"] and a set = {"B", "C", "D"}. Find the common elements (use a set operation: intersection).

my_list = ["A", "B", "C"]
my_set = {"B", "C", "D"}

list_set = set(my_list)
common = list_set.intersection(my_set)
print(f"Intersection of two sets: {common}")
```

```python
# 6. Sukurkite tuple su pelnais kiekvieną mėnesį (12 reikšmių). Išskaičiuokite vidutinį pelną naudodami sum() ir len().
# 6. Create a tuple containing profits for each month (12 values). Calculate the average profit using sum() and len().

profits = (1200, 1500, 1100, 1800, 1700, 1600, 1400, 1550, 1650, 1750, 1900, 2000)
average_profit = sum(profits) / len(profits)

print(f"Average profit: {average_profit}")
```

```python
# 7. Sukurkite dict, kuriame raktais yra mėnesiai (sausis, vasaris...) o reikšmė – pelnas. Apskaičiuokite visų mėnesių sumą.
# 7. Create a dictionary where the keys are months (January, February, etc.) and the values are profits. Calculate the total profit for all months.

profits = {
    "sausis": 1200,
    "vasaris": 1500,
    "kovas": 1100,
    "balandis": 1800,
    "gegužė": 1700,
    "birželis": 1600,
    "liepa": 1400,
    "rugpjūtis": 1550,
    "rugsėjis": 1650,
    "spalis": 1750,
    "lapkritis": 1900,
    "gruodis": 2000
}

total_profit = sum(profits.values())
print(f"Total profit: {total_profit}")
```

```python
from datetime import datetime

# 8. Sukurkite list su pirkimo datomis tekstu (pvz. '2023-01-01') ir konvertuokite jas į datetime objektus naudojant datetime.strptime.
# 8. Create a list of purchase dates as strings (e.g., '2023-01-01') and convert them into datetime objects using datetime.strptime.

dates_str = ["2023-01-01", "2023-02-15", "2023-03-10"]
dates = []

for d in dates_str:
    dates.append(datetime.strptime(d, "%Y-%m-%d"))
    
print(dates_str)
print(dates[-1])
```

```python
# 9. Apskaičiuokite kiek dienų praėjo nuo kiekvienos pirkimo datos iki šiandienos (datetime.date.today()) ir atspausdinkite.
# 9. Calculate how many days have passed from each purchase date until today (datetime.date.today()) and print the results.
from datetime import datetime, date

purchase_dates = [
    datetime.strptime("2022-01-03", "%Y-%m-%d").date(),
    datetime.strptime("2023-04-30", "%Y-%m-%d").date(),
    datetime.strptime("2026-01-25", "%Y-%m-%d").date(),
]

today = date.today()
for d in purchase_dates:
    days_passed = (today - d).days
    print(days_passed)


```

```python
# 10. Sukurkite žodyną, kuriame raktais yra produkto pavadinimai, o reikšmė – sąrašas mėnesinių pardavimų. 
# Raskite bendrą sumą kiekvienam produktui naudodami for ciklą ir sum().

# 10. Create a dictionary where the keys are product names and the values are a list of monthly sales. 
# Using a for loop and sum(), calculate the total sales for each product.

sales = {
    "Bread": [4, 12, 16, 8],
    "Milk": [10, 15, 16, 21],
    "Cheese": [6, 9, 21, 16]
}

for product, monthly_sales in sales.items():
    total = sum(monthly_sales)
    print(f"Product {product} costed in total {total}")
```
