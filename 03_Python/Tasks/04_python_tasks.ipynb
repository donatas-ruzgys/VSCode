```python
from datetime import datetime

# 8. Sukurkite list su pirkimo datomis tekstu (pvz. '2023-01-01') ir konvertuokite jas į datetime objektus naudojant datetime.strptime.
# 8. Create a list of purchase dates as strings (e.g., '2023-01-01') and convert them into datetime objects using datetime.strptime.

dates_str = ["2023-01-01", "2023-02-15", "2023-03-10"]
dates = []

for d in dates_str:
    dates.append(datetime.strptime(d, "%Y-%m-%d"))

print(dates_str)
print(dates[-1])
```

```python
# 9. Apskaičiuokite kiek dienų praėjo nuo kiekvienos pirkimo datos iki šiandienos (datetime.date.today()) ir atspausdinkite.
# 9. Calculate how many days have passed from each purchase date until today (datetime.date.today()) and print the results.
from datetime import date, datetime

purchase_dates = [
    datetime.strptime("2022-01-03", "%Y-%m-%d").date(),
    datetime.strptime("2023-04-30", "%Y-%m-%d").date(),
    datetime.strptime("2026-01-25", "%Y-%m-%d").date(),
]

today = date.today()
for d in purchase_dates:
    days_passed = (today - d).days
    print(days_passed)
```

```python
# 10. Sukurkite žodyną, kuriame raktais yra produkto pavadinimai, o reikšmė – sąrašas mėnesinių pardavimų.
# Raskite bendrą sumą kiekvienam produktui naudodami for ciklą ir sum().

# 10. Create a dictionary where the keys are product names and the values are a list of monthly sales.
# Using a for loop and sum(), calculate the total sales for each product.

sales = {"Bread": [4, 12, 16, 8], "Milk": [10, 15, 16, 21], "Cheese": [6, 9, 21, 16]}

for product, monthly_sales in sales.items():
    total = sum(monthly_sales)
    print(f"Product {product} costed in total {total}")
```

```python
# 1. Parašykite funkciją, kuri priima skaičių ir grąžina tą skaičių padaugintą iš 2.
# 1. Write a function that takes a number and returns that number multiplied by 2.


def multiply_by_two(number):
    return number * 2


a = int(input("Enter a number to be multiplied by two.: "))
result = multiply_by_two(a)
print(f"Number multiplied by two: {result}")
```

```python
# 2. Sukurkite funkciją, kuri grąžina konversijos rodiklį (conversion rate), jei pateikiamas vartotojų skaičius ir pirkėjų skaičius.
# Naudoti: (pirkėjai / vartotojai) * 100.
# 2. Create a function that returns the conversion rate when the number of users and the number of buyers are provided.
# Use: (buyers / users) * 100.


def conversion_rate(users, buyers):
    if users == 0:
        return 0
    return (buyers / users) * 100


rate = conversion_rate(1, 25)
print(rate)
```

```python
# 3. Parašykite funkciją, kuri skaičiuoja CTR (click-through rate): paspaudimai / parodymai * 100.
# 3. Write a function that calculates CTR (click-through rate): clicks / impressions * 100.


def ctr(clicks, impressions):
    if impressions == 0:
        return 0
    return (clicks / impressions) * 100


result = ctr(20, 18)
print(result)
```

```python
# 4. Sukurkite funkciją, kuri apskaičiuoja vidutinę reikšmę iš pateikto sąrašo skaičių. Naudoti sum() ir len().
# 4. Create a function that calculates the average value from a given list of numbers. Use sum() and len().


def average(numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)


result = average([2, 4, 6])
print(result)
```

```python
# 5. Parašykite funkciją, kuri grąžina skirtumą tarp dviejų skaičių – naudinga apskaičiuoti mėnesio pokytį.
# 5. Write a function that returns the difference between two numbers – useful for calculating monthly change.


def difference(current, previous):
    return current - previous


change = difference(40, 70)
print(change)
```

```python
# 6. Sukurkite funkciją, kuri priima pajamų ir išlaidų sąrašus ir grąžina bendrą pelną kiekvienam laikotarpiui (pajamos - išlaidos).
# 6. Create a function that takes lists of revenues and expenses and returns the total profit for each period (revenue − expenses).


def total_profit(revenues, expenses):
    try:
        if len(revenues) != len(expenses):
            raise ValueError("Revenues and expenses lists must be of the same length")

        result = []
        for r, e in zip(revenues, expenses):
            result.append(r - e)

        return result

    except ValueError as error:
        print(error)
        return None


my_revenues = [150, 300, 432, 64]
my_expenses = [345, 15, 389, 33.33]

my_profit = total_profit(my_revenues, my_expenses)
print(my_profit)
```

```python
# 7. Parašykite funkciją, kuri suskaičiuoja, kiek kartų reikšmė 'neaktyvus' yra pateiktame sąraše naudotojų būsenų.
# 7. Write a function that counts how many times the value 'inactive' appears in a given list of user statuses.


def count_inactive(statuses):
    return statuses.count("neaktyvus")


statuses = ["aktyvus", "neaktyvus", "aktyvus", "neaktyvus", "neaktyvus"]
print(count_inactive(statuses))
```

```python
# 8. Parašykite funkciją, kuri priima dict su pardavimais pagal regioną ir grąžina regioną su didžiausia pardavimų suma. Naudoti max().
# 8. Write a function that takes a dictionary of sales by region and returns the region with the highest total sales. Use max().

sales_by_region = {"North": 1200, "South": 950, "East": 1430, "West": 1100}


def best_region(sales):
    return max(sales, key=sales.get)


print(best_region(sales_by_region))
```
