# Paprasti `try / except / finally` pavyzdžiai

Šiame faile – realios situacijos, kuriose dažnai pasitaiko klaidų, ir kaip jas valdyti naudojant `try/except/finally`. Naudojami įrašyti duomenys (be input), bet imituojamos įvesties klaidos.

## 1. Dalyba – kaip išvengti dalybos iš nulio

```python
skaiciai = [(10, 0), (8, 8), (5, 1), (0, 7)]
for a, b in skaiciai:
    try:
        rezultatas = a / b
        print(f"{a} / {b} = {rezultatas}")
    except:
        print(f"Klaida: negalima dalinti {a} iš {b}")
    finally:
        print("--- Baigta viena operacija ---")
```

## 2. Sąrašas su netinkamais duomenimis – tik skaičių suma

```python
duomenys = [10, 20, 'tekstas', 5, None]
suma = 0
for el in duomenys:
    try:
        suma += el
    except:
        print(f"Negalima sudėti: {el}")
print("Galutinė suma:", suma)
```

## 3. Netinkamos operacijos su sąrašu

```python
sarasas = [1, 2, 3, 4, 5, 6]
try:
    print(sarasas[17])
except:
    print("Tokio indekso nėra sąraše")
```

## 4. Skaičių pavertimas – netinkami duomenys tarp gerų

```python
ivestis = ["25", "abc", "14", "-3"]
skaiciai = []
for i in ivestis:
    try:
        skaiciai.append(int(i))
    except:
        print(f"Negalima paversti į skaičių: {i}")
print("Sėkmingai pavertėme:", skaiciai)
```

## 5. `finally` blokas – parodyti, kad kodas veikia toliau

```python
try:
    x = 10 / 0
except:
    print("klaida")
finally:
    print("dirbam")
```
