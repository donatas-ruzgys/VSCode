# Python: `print()` formatavimo pavyzdžiai

## 1. Paprastas `print()` su kintamaisiais

```python
vardas = "Jonas"
amzius = 25
print("Vardas yra: ", vardas)
print("Amžius yra: ", amzius)
```

## 2. Formatavimas naudojant `f-string`

```python
vardas = "Jonas"
amzius = 25
print(f"{vardas} yra {amzius} metų amžiaus")
```

## 3. Formatavimas naudojant `str.format()`

```python
#vardas = "Ona"
#amzius = 29
print("{} yra {} metų amžiaus".format(vardas, amzius))
```

## 4. `sep` panaudojimas (tarpų keitimas)

```python
print("2025", "06", "01", sep=".")
print("Vardas", "Pavardė", sep="_") 
```

## 5. `end` panaudojimas (pabaigos keitimas)

```python
print("Sveiki,", end=" ")
print("kaip sekasi?")
```

## 6. Kelių formatavimo būdų palyginimas

```python
vardas = "Jonas"
amzius = 25

# f-string
print(f"{vardas} - {amzius} m.")

# format()
print("{} - {} m.".format(vardas, amzius))

# klasikinis su kableliais
print(vardas, "-", amzius, "m.")
```
