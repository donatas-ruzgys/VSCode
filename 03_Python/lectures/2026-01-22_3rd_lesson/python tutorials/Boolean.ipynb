# Boolean tipo kintamieji Python kalboje

Loginės reikšmės analizėje: True / False, sąlygos, filtravimas, logikos sprendimai.

## 1. Kas yra Boolean?
- `bool` yra loginis kintamasis: `True` arba `False`
- Gaunamas iš palyginimo, funkcijos, loginės sąlygos

```python
a = 5 > 13
print(a)  # True
```

## 2. Dažniausi palyginimo operatoriai

```python
print(10 == 10)   # True
print(5 != 3)     # True
print(7 < 2)      # False
print(3 >= 3)     # True
```

## 3. Loginiai operatoriai `and`, `or`, `not`

```python
print(True and False)   # False
print(True or False)    # True
print(not True)         # False
print(not False)         # True
```

## 4. Panaudojimas `if` sąlygose

```python
amzius = 200
if amzius >= 18:
    print("Pilnametis")
else:
    print("Nepilnametis")
```

## 5. Boolean funkcija `bool()`

```python
print(bool(None))        # False
print(bool(0))        # False
print(bool([1, 2, "ameeeerika"]))        # True
print(bool(-123))      # True
print(bool(""))       # False
print(bool("begemotas"))   # True
```

## 6. Praktinis pavyzdys: klientų amžiaus tikrinimas

```python
klientai = [15, 18, 21, 16]
for amzius in klientai:
    pilnametis = amzius >= 18
    print(f"Amžius: {amzius}, Pilnametis: {pilnametis}")
```

## 7. Praktinis pavyzdys: ar pirkimo suma viršija ribą

```python
suma = 130
riba = 200
virsijo = suma > riba
if virsijo:
    print("Vartotojas viršijo limitą")
```

## 8. Boolean sąrašų filtravimas (list comprehension)

```python
amziai = [12, 19, 17, 25, 11, 10, 16, 31]
pilnameciai = [a for a in amziai if a >= 18]
print(pilnameciai)
```

## 9. Kelių sąlygų tikrinimas

```python
x = 29
if x > 20 and x < 30:
    print("Tarp 20 ir 30")
```

## 10. Boolean analizė duomenų kontekste

```python
# Pirkimo analizė – ar naudotojas pirko virš 3 kartų ir suma virš 100?
kartai = 4
suma = 150
lojalus = kartai > 3 and suma > 100
print("Lojalus klientas:", lojalus)
```
