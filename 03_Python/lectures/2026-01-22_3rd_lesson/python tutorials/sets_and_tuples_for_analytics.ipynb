# Sets ir Tuples Python kalboje

Demonstraciniai pavyzdžiai, skirti duomenų analitikos praktikai.

## 1. Tuple – fiksuotų reikšmių rinkinys

```python
# Pvz.: kliento duomenų struktūra
klientas = ("Domas", 28, "Vilnius")
vardas, amzius, miestas = klientas
print("Vardas:", vardas)
print("Amžius:", amzius)
print("Miestas:", miestas)
```

## 2. Tuple sąraše – fiksuota struktūra keliose eilutėse

```python
# Pvz.: klientų sąrašas
klientai = [("Jonas", 28), ("Ona", 25), ("Petras", 30)]
for vardas, amzius in klientai:
    print(f"{vardas} yra {amzius} metų.")
```

## 3. Set – unikalių reikšmių rinkinys

```python
# Pvz.: pašalinti dublikatus iš klientų miestų
miestai = ["Vilnius", "Kaunas", "Vilnius", "Klaipėda"]
unikalus_miestai = set(miestai)
print("Unikalūs miestai:", unikalus_miestai)
```

## 4. Set operacijos – grupių palyginimas

```python
# Pvz.: aktyvūs ir registruoti klientai
registruoti = {"Jonas", "Ona", "Petras", "Greta"}
aktyvūs = {"Jonas", "Greta"}

# Kas registruoti, bet neaktyvūs?
neaktyvūs = registruoti - aktyvūs
print("Neaktyvūs klientai:", neaktyvūs)
```

## 5. Set panaudojimas: duomenų analizė

```python
# Pvz.: rasti bendrus produktus dviejose rinkose
lt_produktai = {"obuoliai", "bananai", "kriaušės"}
lv_produktai = {"bananai", "kriaušės", "vynuogės"}

bendri = lt_produktai & lv_produktai
print("Bendri produktai:", bendri)
```

## 6. Tuple kaip rakto panaudojimas žodyne

```python
# Pvz.: pardavimai pagal (regionas, produktas)
pardavimai = {
    ("Vilnius", "obuoliai"): 150,
    ("Kaunas", "obuoliai"): 200,
    ("Vilnius", "kriaušės"): 180
}
print("Pardavimai Vilniuje (obuoliai):", pardavimai[("Vilnius", "obuoliai")])
```

**1. Sankirta (& arba .intersection()): kas bendro**

```python
a = {"A", "B", "C"}
b = {"B", "C", "D"}
print(a & b)  # {'B', 'C'}
print(a.intersection(b))
```

**2. Sąjunga (| arba .union()): visos unikalios reikšmės**
Naudinga sujungiant du skirtingus šaltinius ir norint pašalinti dublikatus.

```python
a = {"A", "B"}
b = {"B", "C"}
print(a | b)  # {'A', 'B', 'C'}
print(a.union(b))
```

**3. Skirtumas (- arba .difference()): kas yra viename, bet ne kitame** Rasti klientus, kurie pirko tik viename regione.

```python
a = {"A", "B", "C"}
b = {"B", "C"}
print(a - b)  # {'A'}
print(a.difference(b))
```

**4. Simetrinis skirtumas (^ arba .symmetric_difference()): kas unikalūs abiejuose, bet ne bendri**
Analizuoti, kurie produktai išskirtiniai kiekvienoje rinkoje.

```python
a = {"A", "B", "C"}
b = {"B", "C", "D"}
print(a ^ b)  # {'A', 'D'}
print(a.symmetric_difference(b))
```

**5. Poaibio tikrinimas (<=)** Patikrinti, ar aktyvūs klientai priklauso visiems registruotiems.

```python
a = {"A", "B"}
b = {"A", "B", "C"}
print(a <= b)  # True
```

**6. Aibės lygybė (==)** Patikrinti, ar dvi klientų grupės visiškai sutampa, nepriklausomai nuo tvarkos.


```python
a = {"A", "B", "O"}
b = {"C", "A", "B"}
print(a == b)  # True
```
