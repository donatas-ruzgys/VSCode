# `datetime` biblioteka Python kalboje

Naudinga darbui su datomis ir laikais – ypač analizuojant laikinius duomenis.

## 1. `datetime.datetime.today()` – dabartinė data ir laikas

```python
import datetime

dabar = datetime.datetime.today()
print("Dabartinė data ir laikas:", dabar)
```

## 2. `datetime.date()` – konkrečios datos objektas

```python
diena = datetime.date(2026, 1, 6)
print("Data:", diena)
```

## 3. `datetime.timedelta()` – laiko skirtumas (pvz., dienų atimtis)

```python
diena1 = datetime.date(2023, 6, 11)
diena2 = datetime.date(2023, 6, 15)

skirtumas = diena2 - diena1
print("Dienų skirtumas:", skirtumas.days)
```

## 4. `strftime()` – datos formatavimas į tekstą

```python
dabar = datetime.datetime.today()
print("Metai-mėnuo-diena:", dabar.strftime("%Y-%m-%d"))
print("Laikas (val:min):", dabar.strftime("%H:%M"))
```

## 5. `strptime()` – konvertavimas iš teksto į datą

```python
tekstas = "2025-03-26"
data = datetime.datetime.strptime(tekstas, "%Y-%m-%d")
print("Gauta data:", data)
```

## 6. Pavyzdys: filtravimas – ar data savaitgalis?

```python
data = datetime.date(2026, 1, 6)
if data.weekday() >= 6:
    print("Tai savaitgalis")
else:
    print("Tai darbo diena")
```

## 7. Generuoti datų seką kas dieną tarp dviejų datų

```python
pradzia = datetime.date(2024, 6, 1)
pabaiga = datetime.date(2024, 6, 11)

diena = pradzia
while diena <= pabaiga:
    print(diena)
    diena += datetime.timedelta(days=1)
```

## 8. Apskaičiuoti vartotojo amžių pagal gimimo datą

```python
gimimo_data = datetime.date(1984, 10, 29)
siandien = datetime.date.today()

amzius = siandien.year - gimimo_data.year
if (siandien.month, siandien.day) < (gimimo_data.month, gimimo_data.day):
    amzius -= 1

print("Amžius:", amzius)
```
