## 5. `scikit-learn` – regresijos modelis

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1], [2], [3]])
y = np.array([2, 4, 6])

model = LinearRegression().fit(X, y)
print("Koeficientas:", model.coef_)
print("Intercepcija:", model.intercept_)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Duomenys
X = np.array([[1], [2], [3]])
y = np.array([2, 4, 6])

# Modelis
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

# Braižymas
plt.scatter(X, y, color='blue', label='Tikri duomenys')
plt.plot(X, y_pred, color='red', label='Regresijos linija')

plt.title("Paprasta tiesinė regresija: y = 2x")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 6. `statsmodels` – regresijos analizė su statistika

```python
import statsmodels.api as sm

X = sm.add_constant([1, 2, 3])
y = [2, 4, 6]

model = sm.OLS(y, X).fit()
print(model.summary())
```

## 7. `datetime` – datų operacijos

```python
import datetime

dabar = datetime.date.today()
print("Šiandien:", dabar)

gimimo_data = datetime.date(1995, 5, 1)
amzius = dabar.year - gimimo_data.year
print("Amžius (metais):", amzius)
```
