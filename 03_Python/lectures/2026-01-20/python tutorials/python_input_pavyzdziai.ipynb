# Python: Pavyzdžiai su `input()`

## 1. Įvestas tekstas

```python
vardas = input("Įveskite savo vardą: ")
print("Labas,", vardas)
```

## 2. Įvestas skaičius

```python
skaicius = int(input("Įveskite skaičių: "))
print("Įvestas skaičius padaugintas iš 2:", skaicius * 2)
```

## 3. Amžiaus tikrinimas

```python
amzius = int(input("Kiek jums metų? "))
if amzius >= 18:
    print("Pilnametis")
else:
    print("Nepilnametis")
```

## 4. Kelių reikšmių įvedimas

```python
vardas = input("Vardas: ")
miestas = input("Miestas: ")
print(vardas, "gyvena mieste", miestas)
```

## 5. Lyginis ar nelyginis

```python
skaicius = int(input("Įveskite skaičių: "))
if skaicius == 0:
    print("Skaičius yra nulis")
elif skaicius % 2 == 0:
    print("Skaičius yra lyginis")
else:
    print("Skaičius yra nelyginis")
```

## 6. Įvedimo patikrinimas

```python
ivestis = input("Įveskite 'taip' arba 'ne': ")
if ivestis.lower() == "taip":
    print("Pasirinkote TAIP")
elif ivestis.lower() == "ne":
    print("Pasirinkote NE")
else:
    print("Nesupratau pasirinkimo")
```
