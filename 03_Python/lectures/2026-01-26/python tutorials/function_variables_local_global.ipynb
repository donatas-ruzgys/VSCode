# Funkcijų kintamieji: Local vs Global (su pavyzdžiais iš duomenų analitikos)

## 1. Lokalus kintamasis (local variable)
- Sukurtas funkcijos viduje
- Naudojamas tik toje funkcijoje

```python
def skaiciuoti_vidurki():
    duomenys = [10, 20, 30]
    vidurkis = sum(duomenys) / len(duomenys)
    print("Vidurkis:", vidurkis)

skaiciuoti_vidurki()
#print(vidurkis)  #  klaida – 'vidurkis' neegzistuoja už funkcijos ribų
```

## 2. Globalus kintamasis – naudojamas (skaitymas) funkcijoje

```python
imones_pavadinimas = "DataSuperPro"

def spausdinti_pavadinima():
    print("Įmonė:", imones_pavadinimas)

spausdinti_pavadinima()
```

## 3. Pakeisti globalų kintamąjį – naudoti `global`

```python
balansas = 1000

def papildyti_balansa(suma):
    global balansas
    balansas += suma
    print("Atnaujintas balansas:", balansas)

papildyti_balansa(500)
print("Galutinis balansas:", balansas)
```

## 4. Geresnė praktika: naudoti `return`, o ne `global`

```python
def papildyti_balansa_teisingai(balansas, suma):
    return balansas + suma

balansas = 1000
balansas = papildyti_balansa_teisingai(balansas, 300)
print("Galutinis balansas:", balansas)
```

## 5. Pavyzdys: konversijos skaičiavimas

```python
def skaiciuoti_konversija(parodymai, paspaudimai):
    konversija = round((paspaudimai / parodymai) * 100,2)
    return konversija

rezultatas = skaiciuoti_konversija(10000, 1450)
print("Konversija: ", rezultatas, "%")
```

## 6. Lokalus kintamasis uždengia globalų (shadowing)

```python
statusas = "Aktyvus"

def keisti_statusa():
    statusas = "Neaktyvus"  # local, neužrašo globalaus
    print("Funkcijoje:", statusas)

keisti_statusa()
print("Globaliai:", statusas)
```
