# Prisijungimas prie Oracle duomenų bazės su Python

Norint prisijungti prie Oracle duomenų bazės Python'e, dažniausiai naudojama biblioteka `cx_Oracle`. Ji leidžia vykdyti SQL užklausas ir dirbti su Oracle DB iš Python programos.

## 1. Būtinos sąlygos

- Įdiegta `cx_Oracle` biblioteka: `pip install cx_Oracle`
- Įdiegti Oracle Instant Client failai (būtini prisijungimui prie Oracle serverio)
- Žinoti prisijungimo duomenis: vartotojo vardą, slaptažodį, host/IP, portą ir servisą

## 2. Prisijungimo pavyzdys

Prisijungimui naudojamas toks šablonas:

```python
import cx_Oracle

dsn = cx_Oracle.makedsn("host", port, service_name="servisas")
connection = cx_Oracle.connect(user="vardas", password="slaptazodis", dsn=dsn)

# Prisijungimas prie Google BigQuery su Python

Norint dirbti su Google BigQuery duomenų baze Python'e, naudojama `google-cloud-bigquery` biblioteka. Ji leidžia autentifikuotis, vykdyti SQL užklausas ir analizuoti duomenis tiesiai iš Jupyter Notebook ar kito Python kodo.

## 1. Būtini paruošiamieji žingsniai

- Turi būti sukurtas Google Cloud projektas su aktyvuotu BigQuery API
- Turi būti sugeneruotas ir atsisiųstas `service account key` failas (`.json`), kuris bus naudojamas autentifikacijai

## 2. Bibliotekos diegimas

```bash
pip install google-cloud-bigquery


````python
from google.cloud import bigquery

client = bigquery.Client()

query = """
SELECT name, SUM(number) as total
FROM `bigquery-public-data.usa_names.usa_1910_current`
WHERE state = 'TX'
GROUP BY name
ORDER BY total DESC
LIMIT 10
"""

df = client.query(query).to_dataframe()
````

````python
# Google Big Query pavyzdys

from google.cloud import bigquery

import pandas as pd

client = bigquery.Client()

query = "SELECT * FROM 'jusu projekto_id.jusudbid.lenteles_pav LIMIT 10'"

query_job = client.query(query)

df = query_job.to_dataframe()

print(df)
````

#### Naudojimas su pandas
Rezultatas tiesiogiai paverčiamas į pandas.DataFrame, todėl toliau galima naudoti groupby(), plot(), merge() ir kt.

BigQuery yra labai galinga debesų analizės platforma, tinkama dirbti su dideliais duomenų kiekiais realiu laiku. Prisijungimas per service account yra standartinis ir saugus būdas dirbti su GCP duomenimis programiškai.
