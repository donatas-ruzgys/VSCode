# Seaborn bibliotekos apžvalga (naudojant *Tips* duomenų rinkinį)

## Kas yra Seaborn
**Seaborn** – tai Python duomenų vizualizacijos biblioteka, sukurta virš `matplotlib`.  
Ji skirta **statistiniams grafiniams vaizdavimams**, glaudžiai integruota su `pandas`.

Pagrindiniai privalumai:
- Automatiškai dirba su `DataFrame`
- Turi estetiškus numatytuosius stilius
- Lengvai kuria statistinius grafikus

---

## Tips duomenų rinkinys
**Tips** – klasikinis demonstracinis `seaborn` duomenų rinkinys apie arbatpinigius restorane.

Pagrindiniai kintamieji:
- `total_bill` – sąskaitos suma
- `tip` – arbatpinigiai
- `sex` – kliento lytis
- `smoker` – ar klientas rūko
- `day` – savaitės diena
- `time` – pietūs arba vakarienė
- `size` – žmonių skaičius prie stalo

---

## Dažniausiai naudojamos Seaborn funkcijos

### Kategoriniai grafikai
Naudojami grupių palyginimui.

- **countplot** – kategorijų dažniai  
- **barplot** – vidurkiai pagal kategorijas  
- **boxplot** – pasiskirstymas ir išskirtys  
- **violinplot** – pasiskirstymo forma  




```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
df.head(30)
```

```python
# correlation with Heatmap
#corelation we find interdependence between all the features.
#it applies to quantitative data only(having either interger or float values)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True);
```

```python
df.corr(numeric_only=True)
```

```python
## Joint Plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='hex')
```

```python
# we can observe that most of the tip is between 2-4USD 
#total bill is in between 10-20USD
```

## Box Plot

```python
sns.boxplot('smoker', 'total_bill',data=df,palette='husl')
```

```python
sns.boxplot('sex', 'total_bill',data=df,palette='Greens')          
```

```python
sns.boxplot('day','total_bill',data=df, palette='Accent')
```

```python
sns.boxplot('day','total_bill',data=df,hue='smoker',palette='Pastel2')
```

```python
# we observe that maximum outliers are for non-smokers on saturday
#sunday smokers we fine no outliers, almost symmetrical boxplot
```

## Violin Plots

```python
## Violin plots helps to see the data both in terms of kernel density estimation
#and box plot
```

```python
sns.violinplot(x='sex', y='total_bill',data=df,palette='rocket')
```

```python
# Majority Males pay bill between 10-20USD. Bill received is maximum in case of males
# Majority Females pay bill between 7-15USD, Bill received has not reached the maximum value
```

```python
sns.violinplot(x='total_bill',y='day',data=df,hue='sex', palette='terrain')
```

```python
# we observe most outliers on saturday. 
#Clearly see how data is distributed or spread.
#most bill range is between 10-20USD
```

```python
sns.violinplot('tip','day', hue='smoker',data=df, palette='seismic')
```

```python
sns.violinplot('total_bill','day',data=df, hue='time',palette='Accent')
```

```python
sns.violinplot('tip','day',palette='Wistia',data=df)
```

```python
sns.violinplot('tip','sex',palette='tab20_r',data=df,hue='time')
```

## Scatter Plots

```python
sns.catplot(x='day',y='tip', data=df)
```

```python
sns.catplot(x='day',y='tip', data=df, jitter=False) 
#notice jitter this reduces the magnitude of jitter in the parameter
```

```python
sns.catplot(x='sex',y='tip',kind='swarm',data=df)
```

We observe that most of the tips are paid between $2-$4 
males paying most $2 tip than females.

```python
sns.catplot(x='sex', y='total_bill', kind='swarm',data=df, palette='husl');
```

```python
sns.catplot(x='sex',y='tip',hue='smoker', data=df, palette='winter',
            kind='swarm')         
```

```python
sns.catplot(x='smoker',y='tip',palette='rainbow',hue='sex',data=df, kind='swarm');
```

we observe that there are more male smokers who pay the tip $2 than non-smokers
for non-smokers also males pay the most tip of $2
we can also infer that males both smoker and non smoker pay the maximum tips.
