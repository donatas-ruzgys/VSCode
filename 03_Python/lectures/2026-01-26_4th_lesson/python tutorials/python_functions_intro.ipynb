## Kas yra funkcija?
Funkcija – tai **pakartotinai naudojamų veiksmų rinkinys**, kuriam galima duoti įvestis (parametrus) ir gauti rezultatą.

### Paprastas pavyzdys

```python
def pasisveikinimas():
    print("Labas, studentai!")

pasisveikinimas()
```

```python
pasisveikinimas()
```

```python
helou = pasisveikinimas
print(helou)
```

### Funkcija su parametrais (argumentais)

```python
def sveikink(vardas):
    print(f"Sveiki, {vardas}!")

sveikink("Aldona")
sveikink("Aloyzas")
```

### Funkcija su grąžinimu (`return`)

```python
def kvadratas(x):
        return x ** 2

rez = kvadratas(17)
print("Skaičiaus kvadratas:", rez)

padauginam = rez * 2
padauginam
```

```python
rezultatas = rez
print(rezultatas)
```

## Pavyzdžiai iš duomenų analitikos

### Konversijos rodiklis

```python
def konversija(pirkimai, apsilankymai):
    if apsilankymai == 0:
        return 0
    return round(pirkimai / apsilankymai, 3)

print("Konversija:", konversija(502, 10003))
```

### PVM skaičiavimas

```python
def skaiciuoti_pvm(suma, tarifas=0.09):
    return round(suma * tarifas, 2)

print("PVM:", skaiciuoti_pvm(2000))
```

## Geroji praktika
- Naudokite reikšmingus funkcijų pavadinimus
- Komentuokite ką funkcija daro
- Naudokite `return`, jei reikia gauti reikšmę
- Venkite perteklinio `print()` funkcijose, nebent tai demonstravimas
