# Python `range()` naudojimas su `for` ir `while` ciklais

## 1. `for` ciklo pavyzdžiai

### 1.1 Skaičiai nuo 0 iki 4

```python
for i in list(range(0,15,3)):
    print(i)
```

### 1.2 Skaičiai nuo 1 iki 5

```python
for i in range(1, 6):
    print(i)
```

### 1.3 Kas antrą skaičių (0, 2, 4...)

```python
for i in range(0, 10, 2):
    print(i)
```

### 1.4 Atgaline tvarka

```python
for i in range(5, 0, -1):
    print(i)
```

## 2. `while` ciklo pavyzdžiai

### 2.1 Paprastas skaičiavimas su `while`

```python
i = 0
while i < 4.9:
    print(i)
    i += 1.1
```

### 2.2 Naudojant `range()` per `list()`

```python
r = list(range(3))
i = 0
while i < len(r):
    print(r[i])
    i += 1
```
