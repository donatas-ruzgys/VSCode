# Python sąrašai (list) – pagrindinės manipuliacijos

## 1. Sąrašo sukūrimas

```python
vaisiai = ["obuolys", "bananas", "kriause"]
print(vaisiai)
```

## 2. Elementų pridėjimas

```python
vaisiai.append("apelsinas")
print(vaisiai)  # Prideda į galą
```

```python
vaisiai.insert(1, "citrina")
print(vaisiai)  # Įterpia į 1 poziciją
```

```python
vaisiai.extend(["vynuoge", "persikas"])
print(vaisiai)  # Prideda kelis
```

## 3. Elementų pašalinimas

```python
vaisiai.remove("bananas")
print(vaisiai)  # Pašalina pagal reikšmę
```

```python
vaisiai.pop(0)
print(vaisiai)  # Pašalina pirmą
```

```python
vaisiai.pop()
print(vaisiai)  # Pašalina paskutinį
```

## 4. Prieiga prie elementų ir jų keitimas

```python
print(vaisiai[0])  # pirmas elementas
```

```python
vaisiai[1] = "kivis"
print(vaisiai)  # keičiamas antras
```

## 5. Paieška ir analizė

```python
print("granatas" in vaisiai)  # ar yra sąraše
```

```python
print(len(vaisiai))  # elementų skaičius
```

```python
print(vaisiai.index("vynuoge"))  # kurioje pozicijoje yra
```

## 6. Rikiavimas ir atvirkštinis tvarkymas

```python
vaisiai.sort()
print(vaisiai)  # abėcėlės tvarka
```

```python
vaisiai.reverse()
print(vaisiai)  # atvirkščiai
```

## 7. Sąrašo išvalymas

```python
vaisiai.clear()
print(vaisiai)  # tuščias sąrašas
```
