# Python žodynai (dictionary) – pagrindinės manipuliacijos

## 1. Žodyno sukūrimas

```python
asmuo = {
    "vardas": "Jonas",
    "amzius": 30,
    "miestas": "Vilnius"
}
print(asmuo)
```

## 2. Prieiga prie reikšmių

```python
print(asmuo["vardas"])
print(asmuo.get("amzius"))
```

## 3. Reikšmės keitimas ir naujo rakto pridėjimas

```python
asmuo["amzius"] = 31
asmuo["el_pastas"] = "jonas@example.com"
print(asmuo)
```

## 4. Raktų, reikšmių ir porų gavimas

```python
print(asmuo.keys())
print(asmuo.values())
print(asmuo.items())
```

## 5. Rakto pašalinimas

```python
asmuo.pop("miestas")
print(asmuo)

del asmuo["vardas"]
asmuo
```

## 6. Tikrinimas ar raktas egzistuoja

```python
print("vardas" in asmuo)
print("miestas" in asmuo)
```

## 7. Elementų skaičius

```python
print(len(asmuo))
```

## 8. Žodyno išvalymas

```python
asmuo.clear()
print(asmuo)
```
