# Prisijungimas prie MySQL su `pymysql` (rekomenduojamas)

Šiame faile naudojamas tik `pymysql`, kuris yra stabilesnis pasirinkimas dirbant su MySQL duomenų baze Jupyter aplinkoje. Vengiame `mysql.connector`, kuris gali sukelti `kernel crash` tam tikrose sistemose.

```python
# !pip install pymysql pandas
import pymysql
import pandas as pd
```

## 1. Prisijungimas prie MySQL ir duomenų nuskaitymas su `pandas.read_sql()`
Prisijungiame prie `sakila` DB ir vykdome paprastą užklausą, kad paimtume aktorių vardus.

```python
try:
    connection = pymysql.connect(
        host="localhost",
        user="root",
        password="5808",
        database="sakila"
    )

    query = "SELECT first_name, last_name FROM actor LIMIT 5"
    df = pd.read_sql(query, connection)
    print(df)
    connection.close()
except Exception as e:
    print("Prisijungimo ar užklausos klaida:", e)
```

## Kodėl `pymysql`?
- Jis yra grynai Python pagrindu – nereikia papildomų DLL failų.
- Stabiliai veikia su `pandas` ir `SQLAlchemy`.
- Nedirba su C-lygio komponentais, todėl rečiau stringa kernel.

Tai yra rekomenduojamas pasirinkimas dirbant su MySQL Jupyter aplinkoje.
