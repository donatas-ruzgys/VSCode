# Python: if, elif, else pavyzdžiai

## 1. Paprastas `if` pavyzdys

```python
amzius = 26

if amzius >= 18:
    print("Jūs esate pilnametis")

```

## 2. `if` / `else` pavyzdys

```python
amzius = 17

if amzius >= 18:
    print("Pilnametis")
else:
    print("Nepilnametis")
```

## 3. `if` / `elif` / `else` – pažymio vertinimas

```python
balas = 4

if balas >= 9:
    print("Puikiai")
elif balas >= 7:
    print("Gerai")
elif balas >= 5:
    print("Patenkinamai")
else:
    print("Nepatenkinamai")
```

## 4. Vartotojo įvestis su tikrinimu

```python
skaicius = int(input("Įveskite skaičių: "))

if skaicius > 0:
    print("Skaičius teigiamas")
elif skaicius < 0:
    print("Skaičius neigiamas")
else:
    print("Skaičius yra nulis")
```

## 5. Sudėtingesnis: kelių sąlygų tikrinimas

```python
vartotojas = input("Įveskite vartotojo vardą: ")
slaptazodis = input("Įveskite slaptažodį: ")

if vartotojas == "admin" and slaptazodis == "1234":
    print("Prisijungimas sėkmingas")
elif vartotojas == "admin":
    print("Neteisingas slaptažodis")
else:
    print("Vartotojas nerastas")
```
