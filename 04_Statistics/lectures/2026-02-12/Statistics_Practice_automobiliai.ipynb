Kaip pasiskirsčiusios kuro sąnaudos? pavaizduokite grafiškai

```python
sns.histplot(df["kuro_sanaudos"], bins=15, kde=True)
plt.title("Kuro sąnaudų pasiskirstymas")
plt.xlabel("Kuro sąnaudos (mpg)")
plt.ylabel("Dažnis")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.histplot(df["kuro_sanaudos"], kde=True, color="green", bins=20)

mean_val = df["kuro_sanaudos"].mean()
median_val = df["kuro_sanaudos"].median()

plt.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color = 'orange', linestyle = '-.', label=f'Median: {median_val:.2f}')

plt.title('Fuel Consumption')
plt.xlabel('Fuel consumption (l/100km)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
